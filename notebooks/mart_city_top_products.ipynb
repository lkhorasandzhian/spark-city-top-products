## Собор витрины mart_city_top_products

### 0. Загрузка данных

```scala
%spark.pyspark

users = spark.createDataFrame(
    [
        ("u1", "Berlin"),
        ("u2", "Berlin"),
        ("u3", "Munich"),
        ("u4", "Hamburg"),
    ],
    ["user_id", "city"]
)

orders = spark.createDataFrame(
    [
        ("o1", "u1", "p1", 2, 10.0),
        ("o2", "u1", "p2", 1, 30.0),
        ("o3", "u2", "p1", 1, 10.0),
        ("o4", "u2", "p3", 5, 7.0),
        ("o5", "u3", "p2", 3, 30.0),
        ("o6", "u3", "p3", 1, 7.0),
        ("o7", "u4", "p1", 10, 10.0),
    ],
    ["order_id", "user_id", "product_id", "qty", "price"]
)

products = spark.createDataFrame(
    [
        ("p1", "Ring VOLA"),
        ("p2", "Ring POROG"),
        ("p3", "Ring TISHINA"),
    ],
    ["product_id", "product_name"]
)

users.show()
orders.show()
products.show()
```

### 1. Подсчёт revenue = qty * price

```scala
%spark.pyspark

orders_r = orders.withColumn("revenue", F.col("qty") * F.col("price"))
orders_r.show()
```

### 2. Объединение orders с users и products

```scala
%spark.pyspark

df_joined = (
    orders_r
    .join(users, on="user_id", how="left")
    .join(products, on="product_id", how="left")
)

df_joined.select(
    "order_id", "user_id", "city", "product_id", "product_name", "qty", "price", "revenue"
).show()
```

### 3. Подсчёт метрики по (city, product_id, product_name):
* orders_cnt (кол-во заказов)
* qty_sum (сумма qty)
* revenue_sum (сумма revenue)

```scala
%spark.pyspark

df_agg = (
    df_joined
    .groupBy("city", "product_id", "product_name")
    .agg(
        F.count("*").alias("orders_cnt"),
        F.sum("qty").alias("qty_sum"),
        F.sum("revenue").alias("revenue_sum"),
    )
)

df_agg.orderBy("city", F.desc("revenue_sum")).show()
```

### 4. Top-2 товара по revenue_sum для каждого города с использованием Window

```scala
%spark.pyspark

w = Window.partitionBy("city").orderBy(F.desc("revenue_sum"), F.desc("qty_sum"), F.asc("product_id"))

df_top2 = (
    df_agg
    .withColumn("rn", F.row_number().over(w))
    .filter(F.col("rn") <= 2)
    .drop("rn")
)

df_top2.orderBy("city", F.desc("revenue_sum")).show()
```

### 5. Запись результата в s3 и в HDFS по пути: /tmp/sandbox_zeppelin/mart_city_top_products/ (parquet, overwrite)

```scala
%spark.pyspark

s3_path = "/tmp/sandbox_zeppelin/s3/mart_city_top_products/"
hdfs_path = "/tmp/sandbox_zeppelin/mart_city_top_products/"

df_top2.write.mode("overwrite").parquet(s3_path)
df_top2.write.mode("overwrite").parquet(hdfs_path)

print("Wrote to:", s3_path)
print("Wrote to:", hdfs_path)
```

### 6. Обратное чтение и демонстрация результата

```scala
%spark.pyspark

df_from_s3 = spark.read.parquet(s3_path)
df_from_hdfs = spark.read.parquet(hdfs_path)
```

```scala
%spark.pyspark

print("From S3:")
df_from_s3.orderBy("city", F.desc("revenue_sum")).show()
```

```scala
%spark.pyspark

print("From HDFS:")
df_from_hdfs.orderBy("city", F.desc("revenue_sum")).show()
```
